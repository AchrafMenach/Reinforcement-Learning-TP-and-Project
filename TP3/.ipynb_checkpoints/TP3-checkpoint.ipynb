```python
import gymnasium as gym
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
env=gym.make("Taxi-v3")
env.close()
```

```python
#Nombre D'etat
state_size=env.observation_space.n
action_size=env.action_space.n

# Création de la table de politique (probabilités égales pour chaque action)
policy_table = np.ones((state_size, action_size)) / action_size

# Création de la table de valeurs initialisée à zéro
value_table = np.zeros(state_size)


print("Les Observation sont :",state_size)
print("Les Action sont :",action_size)#Nombre D'etat
```

```python
# Hyperparamètres optimisés
gamma = 0.99           # Facteur d'actualisation augmenté
lr_policy = 0.001       # Taux d'apprentissage plus agressif
lr_value = 0.1          # Apprentissage valeur plus rapide  
clip_epsilon = 0.2      # Plus grande flexibilité PPO
num_train_episodes = 20000  # Plus long entraînement
num_eval_episodes = 100
entropy_coeff = 0.01    # Moins d'exploration
eps = 1e-8
```

```python
# Initialisation des tables
policy_table = np.ones((state_size, action_size)) / action_size
value_table = np.zeros(state_size)
```

```python
# Initialisation des tables
policy_table = np.ones((state_size, action_size)) / action_size
value_table = np.zeros(state_size)
```

```python
def compute_discounted_rewards(rewards):
    """Calcule les récompenses actualisées cumulées"""
    discounted = np.zeros_like(rewards, dtype=np.float32)
    running_add = 0
    for t in reversed(range(len(rewards))):
        running_add = running_add * gamma + rewards[t]
        discounted[t] = running_add
    return discounted
```

```python
def train_ppo():
    """Entraîne l'agent avec PPO optimisé"""
    global policy_table, value_table
    episode_rewards = []
    success_rates = []
    
    for episode in tqdm(range(num_train_episodes), desc="Entraînement PPO"):
        states, actions, rewards = [], [], []
        state, _ = env.reset()
        done = False
        
        # Collecte des données d'épisode
        while not done:
            # Sélection d'action avec stabilité numérique
            probs = np.clip(policy_table[state], eps, 1-eps)
            probs /= probs.sum()
            action = np.random.choice(action_size, p=probs)
            
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            
            # Reward shaping personnalisé
            if terminated and reward == 20:  # Succès
                reward = 20
            elif terminated:  # Échec
                reward = -10
            else:  # Pas standard
                reward = -1
                
            states.append(state)
            actions.append(action)
            rewards.append(reward)
            state = next_state
        
        episode_rewards.append(sum(rewards))
        
        # Conversion en arrays numpy
        states = np.array(states)
        actions = np.array(actions)
        rewards = np.array(rewards)
        
        # Calcul des récompenses et avantages actualisés
        disc_rewards = compute_discounted_rewards(rewards)
        values = value_table[states]
        advantages = disc_rewards - values
        
        # Normalisation des avantages
        advantages = (advantages - advantages.mean()) / (advantages.std() + eps)
        
        # Mise à jour de la fonction valeur
        value_table[states] += lr_value * (disc_rewards - value_table[states])
        
        # Mise à jour de la politique avec PPO
        old_probs = np.clip(policy_table[states, actions], eps, 1-eps)
        for s, a, adv, old_p in zip(states, actions, advantages, old_probs):
            current_p = np.clip(policy_table[s, a], eps, 1-eps)
            ratio = current_p / old_p
            
            # Calcul de la perte PPO avec clipping
            surr1 = ratio * adv
            surr2 = np.clip(ratio, 1-clip_epsilon, 1+clip_epsilon) * adv
            policy_loss = -np.minimum(surr1, surr2)
            
            # Bonus d'entropie réduit
            entropy = entropy_coeff * np.sum(policy_table[s] * np.log(policy_table[s] + eps))
            
            # Mise à jour
            policy_table[s, a] = np.clip(policy_table[s, a] - lr_policy * (policy_loss - entropy), eps, 1-eps)
            policy_table[s] /= policy_table[s].sum()
        
        # Monitoring périodique
        if episode % 500 == 0:
            test_stats = evaluate_policy("trained", 20)
            success_rates.append(test_stats['success_rate'])
            if test_stats['success_rate'] >= 0.9:  # Early stopping
                print(f"\nEarly stopping à l'épisode {episode}, taux de succès: {test_stats['success_rate']:.0%}")
                break
    
    return episode_rewards, success_rates

def evaluate_policy(policy, num_episodes, render=False):
    """Évalue une politique sur plusieurs épisodes"""
    rewards = []
    lengths = []
    successes = 0
    
    for ep in range(num_episodes):
        state, _ = env.reset()
        done = False
        ep_reward = 0
        
        while not done:
            if render and ep == 0:
                env.render()
                
            if policy == "random":
                action = env.action_space.sample()
            else:
                action = np.argmax(policy_table[state])
            
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            ep_reward += reward
            state = next_state
        
        rewards.append(ep_reward)
        lengths.append(len(lengths))
        if ep_reward > 0:
            successes += 1
    
    stats = {
        'mean_reward': np.mean(rewards),
        'std_reward': np.std(rewards),
        'success_rate': successes / num_episodes,
        'mean_length': np.mean(lengths),
    }
    return stats
```

```python
def evaluate_policy(policy, num_episodes, render=False):
    """Évalue une politique sur plusieurs épisodes"""
    rewards = []
    lengths = []
    successes = 0
    
    for ep in range(num_episodes):
        state, _ = env.reset()
        done = False
        ep_reward = 0
        
        while not done:
            if render and ep == 0:
                env.render()
                
            if policy == "random":
                action = env.action_space.sample()
            else:
                action = np.argmax(policy_table[state])
            
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            ep_reward += reward
            state = next_state
        
        rewards.append(ep_reward)
        lengths.append(len(lengths))
        if ep_reward > 0:
            successes += 1
    
    stats = {
        'mean_reward': np.mean(rewards),
        'std_reward': np.std(rewards),
        'success_rate': successes / num_episodes,
        'mean_length': np.mean(lengths),
    }
    return stats
```

```python
def plot_results(train_rewards, success_rates):
    """Visualisation des résultats"""
    plt.figure(figsize=(15, 5))
    
    # Récompenses d'entraînement
    plt.subplot(1, 2, 1)
    window = 100
    moving_avg = np.convolve(train_rewards, np.ones(window)/window, mode='valid')
    plt.plot(train_rewards, alpha=0.2, label='Par épisode')
    plt.plot(range(window-1, len(train_rewards)), moving_avg, 'r-', label=f'Moyenne {window} épisodes')
    plt.title('Récompenses pendant l\'entraînement')
    plt.xlabel('Épisodes')
    plt.ylabel('Récompense')
    plt.legend()
    plt.grid(True)
    
    # Taux de succès
    plt.subplot(1, 2, 2)
    plt.plot(np.linspace(0, len(train_rewards), len(success_rates)), success_rates, 'g-')
    plt.title('Évolution du taux de succès')
    plt.xlabel('Épisodes')
    plt.ylabel('Taux de succès')
    plt.ylim(0, 1)
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()

# Évaluation avant entraînement
print("=== Évaluation avant entraînement ===")
random_stats = evaluate_policy("random", num_eval_episodes)
print(f"Récompense moyenne: {random_stats['mean_reward']:.2f} ± {random_stats['std_reward']:.2f}")
print(f"Succès: {random_stats['success_rate']:.0%}")

# Entraînement
train_rewards, success_rates = train_ppo()

# Évaluation après entraînement
print("\n=== Évaluation après entraînement ===")
trained_stats = evaluate_policy("trained", num_eval_episodes, render=True)
print(f"Récompense moyenne: {trained_stats['mean_reward']:.2f} ± {trained_stats['std_reward']:.2f}")
print(f"Succès: {trained_stats['success_rate']:.0%} | Longueur moyenne: {trained_stats['mean_length']:.1f}")

# Visualisation
plot_results(train_rewards, success_rates)
env.close()
```
